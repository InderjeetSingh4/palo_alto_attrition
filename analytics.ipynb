```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Palo Alto Networks.csv')
```

```python
df.head()
```

```python
dept_attrition = df.groupby('Department')['Attrition'].agg(['sum', 'count'])
dept_attrition['Attrition_Rate_Pct'] = (dept_attrition['sum'] / dept_attrition['count']) * 100
print(dept_attrition)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=dept_attrition.index, y=dept_attrition['Attrition_Rate_Pct'], palette="viridis")
plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

```python
ot_attrition = df.groupby('OverTime')['Attrition'].mean() * 100
plt.figure(figsize=(6, 5))
sns.barplot(x=ot_attrition.index, y=ot_attrition.values, palette="mako")
plt.title("Impact of Overtime on Attrition")
plt.ylabel("Attrition Rate (%)")
plt.show()
```
